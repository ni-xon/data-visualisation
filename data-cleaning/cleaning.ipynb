# Astronauts Dataset Wrangling

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("astronauts.csv")
```

## Creating nationalities.csv
Distribution of astronauts nationalities across the world

### Replacing country names to match world.topojson field values

```python
raw_data['nationality'].replace({'U.S.': 'United States', 'UAE': 'United Arab Emirates', 
                                 'U.S.S.R/Ukraine': 'Ukraine', 'U.K.': 'United Kingdom', 
                                 'Hungry': 'Hungary', 'Netherland': 'Netherlands', 
                                 'Republic of South Africa': 'South Africa', 'Czechoslovakia': 'Czech Rep.'}, inplace=True)
```

```python
raw_data.head()
```

### 

### Creating df for unique astronauts

```python
astronauts_unique_df = raw_data.drop_duplicates(subset=['name'])
```

```python
astronauts_unique_df
```

```python
astronauts_unique_df['nationality'].unique()
```

### Identifying number of people with dual citizenship

```python
astronauts_unique_df[astronauts_unique_df['nationality'] == 'U.K./U.S.']
```

```python
temp = astronauts_unique_df['nationality'].value_counts().reset_index()
```

```python
temp.rename(columns = {'index': 'country', 'nationality': 'count'}, inplace=True)
```

```python
temp
```

```python
# Delete rows with U.S./U.K.
temp = temp.drop([8]).reset_index().drop(columns=['index'])
```

```python
temp
```

### Adjust for dual citizenship U.S./U.K.

```python
# Add counts to US and UK individually
temp.at[0, 'count'] += 3 # US
temp.at[10, 'count'] += 3 # UK
```

```python
temp.to_csv('nationalities.csv', index=False)
```

## Creating genders.csv
Distribution of genders

```python
sex_df = astronauts_unique_df['sex'].value_counts().reset_index()
```

```python
sex_df.rename(columns={'index': 'sex', 'sex': 'count'})
```

```python
sex_df.to_csv('sex_df.csv', index=False)
```

## Creating occupation.csv

Roughly 60% of astronauts selected by NASA have come from military service.

```python
military_civilian_df = astronauts_unique_df['military_civilian'].value_counts().reset_index()
```

```python
military_civilian_df.rename(columns={'index': 'military_civilian', 'military_civilian': 'count'})
```

```python
military_civilian_df.to_csv('military_civilian_df.csv', index=False)
```
